# Terrain Explorer

Interactive terrain exploration with datasets, `ViewerWidget`, and programmatic camera control.

```python
import forge3d as f3d

dem_path = f3d.fetch_dem("rainier")
widget = f3d.ViewerWidget(terrain_path=dem_path, src=f3d.mini_dem_path(), width=1100, height=700)
widget
```

## Camera and sun presets

```python
widget.set_camera(phi_deg=20, theta_deg=52, radius=5400, fov_deg=42)
widget.set_sun(azimuth_deg=300, elevation_deg=26)
widget.snapshot("rainier-overview.png")
```

## Overlay workflow

The same widget can control raster overlays through the viewer handle.

```python
overlay_path = f3d.fetch_dataset("swiss-land-cover")
widget.send_ipc({
    "cmd": "load_overlay",
    "name": "land-cover",
    "path": str(overlay_path),
    "opacity": 0.8,
    "z_order": 10,
})
```

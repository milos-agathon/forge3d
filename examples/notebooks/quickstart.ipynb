# forge3d Quickstart

A minimal path from packaged sample data to a live viewer or notebook preview.

```python
import forge3d as f3d

print(f3d.available_datasets())
print(f3d.dem_stats(f3d.mini_dem()))
```

## Notebook preview

`ViewerWidget` tries to launch the Rust viewer and falls back to `RenderView` when needed.

```python
widget = f3d.ViewerWidget(
    terrain_path=f3d.mini_dem_path(),
    src=f3d.mini_dem_path(),
    width=960,
    height=600,
)
widget
```

## Scripted desktop viewer

Run this cell after building `interactive_viewer` with `cargo build --release --bin interactive_viewer`.

```python
with f3d.open_viewer_async(terrain_path=f3d.mini_dem_path(), width=1400, height=900) as viewer:
    viewer.set_orbit_camera(phi_deg=35, theta_deg=55, radius=1.8, fov_deg=45)
    viewer.set_sun(azimuth_deg=315, elevation_deg=32)
    viewer.snapshot("mini-dem.png")
```
